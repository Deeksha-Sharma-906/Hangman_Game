```python
import random
```

```python
stages =[ '''
 +---+
 |   |
 O   |
/|\  |
/ \  |
     |
======
''','''
 +---+
 |   |
 O   |
/|\  |
/    |
     |
======
''','''
 +---+
 |   |
 O   |
/|\  |
     |
     |
======
''','''
 +---+
 |   |
 O   |
/|   |
     |
     |
======
''','''
 +---+
 |   |
 O   |
 |   |
     |
     |
======
''','''
 +---+
 |   |
 O   |
     |
     |
     |
======
''','''
 +---+
 |   |
     |
     |
     |
     |
======
''']
```

```python
word_list = ["mumbai", "bangalore", "pune", "hyderabad", "delhi"] #words list for trial

#Choose a random word from the list of words
chosen_word = random.choice(word_list)

placeholder =""

#creating blanks to guess the word
for position in range(len(chosen_word)):
    placeholder+= "_ "
print(placeholder)

#number of lives in the game
lives = 6

game_over =False

correct_word = []

while not game_over:
    print(f"\n********************{lives}/6 lives left********************")
    choice = input("Guess a letter: ").lower() #input from user to guess the letter
    print(f"You have guessed {choice} letter")    
    
    display =""
    #if guessed letter is correct
    for letter in chosen_word:
        if letter == choice:
            correct_word.append(letter) #if letter guessed by user is correct, append it in the list
            display += letter
        elif letter in correct_word:
            display += letter
        else:
            display +="_ "
    
    print(display)
    #if guessed letter is incorrect
    if choice not in chosen_word:
        lives -=1
        print(f"Incorrect choice of {choice} so you lost a chance.")
        if lives ==0:
            game_over=True
            show = chosen_word.upper()
            print(f"\n********************IT WAS {show}. YOU LOSE********************")

    if "_" not in display:
        game_over = True
        print("\n********************HURRAYYY!!! YOU WIN!!!********************")

    print(stages[lives]) #display the current stage of the gamer
```
